- After thorough research, I found out that NaN stands for "No Medal".
- According to https://olympics.fandom.com/wiki/Olympics_Wiki, 
all the players that have NaN values in Medal column, have not won any medal.
- So we can assume that NaN stands for "No Medal".

```python
# replace the missing values in Medal to "No Medal"
df_athlete["Medal"].fillna("No Medal", inplace=True)
df_athlete
```
