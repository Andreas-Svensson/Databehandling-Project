```python
import pandas as pd


Athlete_events = pd.read_csv("Data/athlete_events.csv")
Noc_regions = pd.read_csv("Data/noc_regions.csv")

df_athlete = pd.DataFrame(Athlete_events)
df_regions = pd.DataFrame(Noc_regions)
```

## Uppgift 1 - Landstatistik


Ni får ett land tilldelat till er grupp. Börja med att anonymisera kolumnen med idrottarnas namn med
hashfunktionen SHA-256

```python
df_athlete
```

```python
import hashlib as hl


#creates hash unsure what type 
#---------------------------------------------------------------------first method i used
# df_athlete["Name"] = df_athlete["Name"].apply(hash)---- this works 
#----------------------------------------------------------------------

# This works aswell

df_athlete["Name"] = df_athlete["Name"].apply(
    lambda x: 
        hl.sha256(x.encode()).hexdigest()
)


df_athlete

```

```python
df_athlete.to_csv("Data/anonymized_olympics_data.csv") # saving anonymised data to file to use instead of raw data
```

```python
## creates new dataframe with only untied states

#creates condition
country_condition = ['United States']

df_USA = df_athlete[df_athlete['Team'].isin(country_condition)] ## apply condition

df_USA #created dataframe with only american teams
```

```python
import plotly_express as px

#groups and count

# from df_USA groups  sport medal and sort by medal
medal_USA = df_USA[['Sport','Medal']].groupby("Sport").count().reset_index().sort_values(by="Medal", ascending= False)


fig = px.bar(medal_USA, x= "Sport", y="Medal", log_y=True) #making logaretmic might be confusing 

fig.show()
```

```python
winter_con = ['Winter'] ##creates keyword

df_winter = df_USA[df_USA["Season"].isin(winter_con)] # apply keyword

#groups count and storts by medal
df_winter = df_winter[["Sport", "Medal"]].groupby("Sport").count().reset_index().sort_index().sort_values(by="Medal", ascending= False)


fig = px.bar(df_winter, x= "Sport", y="Medal", log_y=True) #making logaretmic might be confusing 

fig.show()
```

```python
summer_con = ["Summer"] #same logic df_winter and df_usa

df_summer = df_USA[df_USA["Season"].isin(summer_con)]

df_summer = df_summer[["Sport", "Medal"]].groupby("Sport").count().reset_index().sort_index().sort_values(by="Medal", ascending= False)



fig = px.bar(df_summer, x= "Sport", y="Medal", log_y=True) #making logaretmic might be confusing 

fig.show()
```

```python
women_con = ['F']

df_age = df_USA[df_USA["Sex"].isin(women_con)]


fig = px.histogram(df_age, x="Age",  log_y=True, title="Women age distribution")

fig.show()
# summer_con = ["Summer"] #same logic df_winter and df_usa

# df_summer = df_USA[df_USA["Season"].isin(summer_con)]
```

```python
men_con = ['M']

df_age = df_USA[df_USA["Sex"].isin(men_con)]


fig = px.histogram(df_age, x="Age",  log_y=True, title=" age distribution")

fig.show()
```

```python
#todo 

## create df with only mean age and medal by grouby 

df_medel_age_mean = df_age[["Age", "Medal"]].groupby("Medal").mean().reset_index().sort_index().sort_values(by="Medal")


df_medel_age_mean

fig = px.bar(df_medel_age_mean, x= "Medal", y="Age", title = "American median age winners") 

fig.show()
```

```python
df_medel_age_max = df_age[["Age", "Medal"]].groupby("Medal").max().reset_index().sort_index().sort_values(by="Medal")


df_medel_age_max

fig = px.bar(df_medel_age_max, x= "Medal", y="Age", title=("Oldest American winners")) 

fig.show()
```

```python
df_medel_age_min = df_age[["Age", "Medal"]].groupby("Medal").min().reset_index().sort_index().sort_values(by="Medal")


df_medel_age_min

fig = px.bar(df_medel_age_min, x= "Medal", y="Age", title=("Youngest American winners")) 
fig.show()
```
