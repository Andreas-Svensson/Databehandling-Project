```python
# import regex as re

df_un_list = pd.read_html("https://www.worlddata.info/alliances/un-united-nations.php")[0]
# df_un_list[0]

df_un_list
```

```python

df_un_list['Country'] = df_un_list['Country'].replace('Antigua and Barbuda', 'Antigua')

df_un_list['Country'] = df_un_list['Country'].replace('Burma', 'Myanmar')

df_un_list['Country'] = df_un_list['Country'].replace('Central Africa', 'Central African Republic')

df_un_list['Country'] = df_un_list['Country'].replace('Congo', 'Republic of Congo') 

df_un_list['Country'] = df_un_list['Country'].replace('Congo (Dem. Republic)', 'Democratic Republic of the Congo') 

df_un_list['Country'] = df_un_list['Country'].replace('Eswatini', 'Swaziland')


df_un_list['Country'] = df_un_list['Country'].replace('North Macedonia', 'Macedonia')


df_un_list['Country'] = df_un_list['Country'].replace('Tuvalu', 'Tuvalu')
df_un_list['Country'] = df_un_list['Country'].replace('Trinidad and Tobago', 'Trinidad')


df_un_list['Country'] = df_un_list['Country'].replace('United Kingdom', 'UK')

df_un_list['Country'] = df_un_list['Country'].replace('Bolivia', 'Boliva')



df_un_list['Country'] = df_un_list['Country'].replace('Saint Kitts,Turks and Caicos Islands', 'Saint Kitts,Turks and Caicos Islands')

len(set(df_un_list["Country"]).intersection(set(df_regions['region'])))

# df_un_list.drop(["Commencement", "Population", "Area" ], axis =1)
#test 

df_un_list.drop(["Commencement",  "Population", "Area"], axis=1, inplace= True)
```
